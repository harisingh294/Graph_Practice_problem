```python
def Check_Complete(Adj_matrix):
    # Get the number of nodes in the graph (dimensions of the adjacency matrix)
    dimensions = len(Adj_matrix)  

    # Check if each row has the correct number of elements
    for row in Adj_matrix.values():
        if len(row) != dimensions:
            return "Enter a valid adjacency matrix"

    # Check each element in the adjacency matrix
    for i in range(dimensions):
        for j in range(dimensions):
            current_node = list(Adj_matrix.keys())[i]  # Get the node name at index i
            
            # Diagonal elements must be 0 (no self-loops in a complete graph)
            if i == j:
                if Adj_matrix[current_node][j] != 0:
                    return "Not a complete graph"
            else:
                # Off-diagonal elements must be 1 (each node is connected to every other node)
                if Adj_matrix[current_node][j] != 1:
                    return "Not a complete graph"

    # If all conditions are satisfied, the graph is complete
    return "Complete graph"

# Test case with a 3-node complete graph
result = Check_Complete({
    'ch': [0, 1, 1],
    'hg': [1, 0, 1],
    'hr': [1, 1, 0]
})

print(result)
```

```python
# ----------- Fresh Adjacency List with Unique Names ----------- #
adjacency_list = {
    "Aarav": ["Kiara", "Dev"],
    "Kiara": ["Aarav", "Meera"],
    "Dev": ["Aarav", "Ishaan"],
    "Meera": ["Kiara"],
    "Ishaan": ["Dev"]
}

# ----------- Function to Check if Graph is Connected ----------- #
def is_graph_connected(graph):
    visited = []              # To track visited nodes
    nodes = list(graph)       # Get all node names

    # Simple Depth First Search (DFS)
    def dfs(node):
        visited.append(node)
        for neighbor in graph[node]:
            if neighbor not in visited:
                dfs(neighbor)

    # Start DFS from the first node
    dfs(nodes[0])

    # Check if all nodes were visited
    for node in nodes:
        if node not in visited:
            return False
    return True

# ----------- Run the Check ----------- #
if is_graph_connected(adjacency_list):
    print("The graph is connected.")
else:
    print("The graph is not connected.")
```

```python
# Sample graph (undirected)
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'C', 'D'],
    'C': ['A', 'B', 'D'],
    'D': ['B', 'C', 'E'],
    'E': ['D']
}

# Sample vertex sequence (you can change this to test others)
vertex_sequence = ['A', 'B', 'C', 'D', 'E']

# Function to check walk/trail/path
def classify_sequence(graph, sequence):
    if len(sequence) < 2:
        return "Too short to classify."

    used_edges = set()
    visited_vertices = set()

    is_walk = True
    is_trail = True
    is_path = True

    for i in range(len(sequence) - 1):
        u, v = sequence[i], sequence[i + 1]

        # Check if there's an edge between u and v
        if v not in graph.get(u, []):
            is_walk = False
            break

        # Check if edge was used before (for trail)
        edge = tuple(sorted((u, v)))
        if edge in used_edges:
            is_trail = False
        else:
            used_edges.add(edge)

        # Check if vertex is repeated (for path)
        if v in visited_vertices:
            is_path = False
        else:
            visited_vertices.add(v)

    if not is_walk:
        return "Not a walk."
    elif is_walk and not is_trail:
        return "It's a walk."
    elif is_trail and not is_path:
        return "It's a trail."
    elif is_path:
        return "It's a path."

# Run the check
result = classify_sequence(graph, vertex_sequence)
print("Sequence:", " → ".join(vertex_sequence))
print("Result:", result)
```

```python
def is_tree(graph):
    visited = set()

    def dfs(node, parent):
        visited.add(node)
        for neighbor in graph[node]:
            if neighbor == parent:
                continue
            if neighbor in visited or not dfs(neighbor, node):
                return False
        return True

    # Start DFS from an arbitrary node (first key)
    start_node = next(iter(graph))
    if not dfs(start_node, None):
        return False

    # Check if all nodes were visited (graph is connected)
    return len(visited) == len(graph)
graph = {
    0: [1, 2],
    1: [0, 3],
    2: [0],
    3: [1]
}

print(is_tree(graph))
```

```python
# Function to build an undirected graph using adjacency list representation
def build_graph(edges, num_nodes):
    graph = []

    # Initialize the graph with empty lists for each node
    for _ in range(num_nodes):
        graph.append([])

    # Add edges to the graph (bidirectional since it's an undirected tree)
    for i in range(len(edges)):
        u = edges[i][0]
        v = edges[i][1]
        graph[u].append(v)
        graph[v].append(u)
    
    return graph

# Function to find the height of the tree using DFS
def find_height(graph, node, parent):
    max_height = 0  # Track the maximum height among all children

    # Visit all neighbors (children) of the current node
    for i in range(len(graph[node])):
        neighbor = graph[node][i]
        
        # Avoid going back to the parent (prevent infinite loop in undirected graph)
        if neighbor != parent:
            h = find_height(graph, neighbor, node)  # Recursive DFS call
            if h > max_height:
                max_height = h  # Update max height if a taller subtree is found

    return max_height + 1  # Add 1 to include current level

# Define the edges of the tree
edges = [
    (0, 1),
    (0, 2),
    (2, 3),
    (2, 4),
    (4, 5)
]

num_nodes = 6  # Total number of nodes in the tree

# Build the graph from the edge list
graph = build_graph(edges, num_nodes)

# Compute the height of the tree from root node 0
# Subtract 1 to count edges instead of nodes in the height
height = find_height(graph, 0, -1) - 1

# Output the height
print("Height of the tree:", height)
```

```python
def build_graph(edges, num_nodes):
    graph = []
    for _ in range(num_nodes):
        graph.append([])
    
    for i in range(len(edges)):
        u = edges[i][0]
        v = edges[i][1]
        graph[u].append(v)
        graph[v].append(u)

    return graph

def compute_depths(graph, node, parent, depth, depths):
    depths[node] = depth
    for i in range(len(graph[node])):
        neighbor = graph[node][i]
        if neighbor != parent:
            compute_depths(graph, neighbor, node, depth + 1, depths)

# Define edges
edges = [
    (0, 1),
    (0, 2),
    (2, 3),
    (2, 4),
    (4, 5)
]

num_nodes = 6
graph = build_graph(edges, num_nodes)

# To store depth of each node
depths = []
for _ in range(num_nodes):
    depths.append(0)

# Start DFS from root (node 0)
compute_depths(graph, 0, -1, 0, depths)

# Output depths
print("Node depths:")
for i in range(num_nodes):
    print("Node", i, "-> Depth:", depths[i])

# Tree depth (max depth)
max_depth = 0
for i in range(num_nodes):
    if depths[i] > max_depth:
        max_depth = depths[i]

print("Depth of the tree:", max_depth)
```
